```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import MinMaxScaler,PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```
_df = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter07/Dataset/superconduct/train.csv')
```

```
X=_df.drop('critical_temp',axis=1).values
y=_df[['critical_temp']].values
```

```
X_train,X_eval,y_train,y_eval=train_test_split(X,y,train_size=0.8,random_state=0)
```

```
lr_model1=LinearRegression()
```

```
lr_model1.fit(X_train,y_train)
```

```
lr_model1_preds=lr_model1.predict(X_eval)
```

```
from sklearn.metrics import mean_squared_error
```

```
print('R2 for lr_model1:{}'.format(lr_model1.score(X_eval,y_eval)))
print('MSE for lr_model1:{}'.format(mean_squared_error(y_eval,lr_model1_preds)))
```

```
steps=[('scaler',MinMaxScaler()),
       ('poly',PolynomialFeatures(degree=3,interaction_only=True)),
       ('lr',LinearRegression())]
```

```
lr_model2=Pipeline(steps)
```

```
lr_model2.fit(X_train,y_train)
```
